## Smart Cities

Universidade de Vienas em 2007, organizou as 6 reas que compem uma Smart Cities:

- <b>Smart People</b>

- <b>Smart Living</b>, como as casas dialogam com a cidade

- <b>Smart Mobility</b>, trnsito e mobilidade

- <b>Smart Environment</b>, questo de recursos naturais como agua, alimentao e energia para estas reas

- <b>Smart Governance</b>, como os cidades tomam suas decisees e como politica pode ser feita de forma clara, como banco de dados abertos a populao

- <b>Smart Economy</b>, mas aproveitar o sistema de forma sustentvel de forma criativa

## Smart People

Caracteristicas que definem Smart People:

- Sociedade com Educao de qualidade

- Sociedade com Incluso

- Sociecadae com Diversidade

- Resultado dos outros itens, indica tambm maior Expectativa de Vida, por estarem motivadas

## Smart Economy

Caracteristicas que definem Smart Economy:

- Flexibilidade no mercado de trabalho

- Integrao internacional, que receba novos cidades internacional, tanto para incorporar o mercado local, quanto para auxiliar pessoas que buscam auxilio

- Espirito de inovao

- Empreendedorismo

- Produtividade

- Transformao, para que a economia se renove sempre, diferente do que ocorreu com Detroit, na qual no estimularam novas formar de emprego (alm do setor automobilista), com isso a cidade teve uma grande queda na qual no conseguiu se sustentar. Por tanto apoie novos projetos, principalmente os sustentveis


## Smart Environment

Caracteristicas que definem Smart Environment:

- Proteo ambiental

- Administrao dos recursos sustentveis

- Aproveitar dos recursos naturais em prol da cidade, sempre com sustentabilidade

- Qualidade do ar

Exemplo:

Projeto do governo coreano, junto a empresas que investiram 40 Bilhes de dolares, para construir uma cidade que promete ser o polo de tecnologia e finanas da sia, um exemplo disto  o mega projeto de limpeza de rios da capital Seul.

Outro exemplo deste local  que em Songdo, todas as liveira so conectadas  central de cooleta de livo via vcuo, alm de um programa de pontuao para os cidadees que depositam seu lixo de maneira correta nas lixeiras de reciclagem.

## Smart Governance

Principais caracteristica que definem Smart Governance:

- Eficincia e transparncia na administrao

- Servios pblicos e com preocupao social

- Perspectiva e estratgia polticas
