## Welcome to the smart cities

Aproveitar os recursos com sustentabilidade e inteligencia

## Smart Cities

Universidade de Vienas em 2007, organizou as 6 áreas que compõem uma Smart Cities:

- <b>Smart People</b>

- <b>Smart Living</b>, como as casas dialogam com a cidade

- <b>Smart Mobility</b>, trânsito e mobilidade

- <b>Smart Environment</b>, questão de recursos naturais como agua, alimentação e energia para estas áreas

- <b>Smart Governance</b>, como os cidadões tomam suas decisõees e como politica pode ser feita de forma clara, como banco de dados abertos a população

- <b>Smart Economy</b>, mas aproveitar o sistema de forma sustentável de forma criativa

## Smart People

Caracteristicas que definem Smart People:

- Sociedade com Educação de qualidade

- Sociedade com Inclusão

- Sociecadae com Diversidade

- Resultado dos outros itens, indica também maior Expectativa de Vida, por estarem motivadas

## Smart Economy

Caracteristicas que definem Smart Economy:

- Flexibilidade no mercado de trabalho

- Integração internacional, que receba novos cidadões internacional, tanto para incorporar o mercado local, quanto para auxiliar pessoas que buscam auxilio

- Espirito de inovação

- Empreendedorismo

- Produtividade

- Transformação, para que a economia se renove sempre, diferente do que ocorreu com Detroit, na qual não estimularam novas formar de emprego (além do setor automobilista), com isso a cidade teve uma grande queda na qual não conseguiu se sustentar. Por tanto apoie novos projetos, principalmente os sustentáveis


## Smart Environment

Caracteristicas que definem Smart Environment:

- Proteção ambiental

- Administração dos recursos sustentáveis

- Aproveitar dos recursos naturais em prol da cidade, sempre com sustentabilidade

- Qualidade do ar

Exemplo:

Projeto do governo coreano, junto a empresas que investiram 40 Bilhões de dolares, para construir uma cidade que promete ser o polo de tecnologia e finanças da Ásia, um exemplo disto é o mega projeto de limpeza de rios da capital Seul.

Outro exemplo deste local é que em Songdo, todas as liveira são conectadas á central de cooleta de livo via vácuo, além de um programa de pontuação para os cidadõees que depositam seu lixo de maneira correta nas lixeiras de reciclagem.

## Smart Governance

Principais caracteristica que definem Smart Governance:

- Eficiência e transparência na administração

- Serviços públicos e com preocupação social

- Perspectiva e estratégia políticas

## Smart Living

Não é apenas uam casa cheia de sensores ou conceito de IoT, é preciso algo além disto, devemos pensar no macro, na cidade na sociedade, como as principais caracteristicas de uma Smart Living:

- Opções de lazer e culturais

- Acesso á sáude

- Segurança

- Qualidade habitacional

- Acesso á educação de qualidade

- Atrações turísticas

## Smart Mobility

As principais caracteristicas são:

- Sistema de transporte público acessível e de qualidade

- Acessibilidade

- Infraestrutura baseada em Tecnologia da Informação e Comunicação

- Sustentabilidade no sistema de transporte
