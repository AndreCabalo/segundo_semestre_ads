## Ios

- 30% dos usurios utilizam (em 2023)

- Precisa de um equipamento Apple para publicas alm da exigencia da taxa de desenvolvedor para publicao $99,00/ano, para entrar na loja

### Primeiros passos com Ios

1 -  Baixar Ebook do Swift para aprender o basico

2 - Usar um software (Por exemplo: REPL - Read-Eval-Print Loop) para entender a dinamica da linguagem,e testar para criar classse, chamar o mtodo 

3 - Se gostar, utilize um Mac na MacCloud.com, onde tera uma mquina mac, para usar o XCode

4 - Interagir com varias telas, interagir model com controler e view

5 - Acesso a Json

6 - Persistencia de dados

## Android

- Taxa para confirmao de que  um ser humano de R$ 20,00 para confirmaoo do desenvolvedor

### Primeiro passos com Android:

- Baixar o Android Studio (um set de ferramentas), apos isso:
