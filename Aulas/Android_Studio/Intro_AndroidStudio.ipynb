# A Casa do Android


- Uma das ferramentas mais importantes para desenvolvimento Android é a IDE Android Studio, desenvolvi pela JetBrain e indicado pelo Google para desenvolvimento de app Android.

    1- Primeiro precisaremos baixar o Android Studio, em: <a href="https://developer.android.com/studio"> Studio Android</a>

    2- Em seguida precisamos também que esteja instalado o JDK do Java, pois o Android Studio dependa do JDK do Java. por tanto baixe este atráves do link: https://www.oracle.com/br/java/technologies/downloads/

    3- Após o Download de ambos, primeiro instale o JDK do Java.

    4- Na sequência o Android Studio, mantendo a opção de Android Virtual Device (também).

    5- Pronto, feito isso, vc ja tera o Android Studio rodando em sua maquina.

## Criando o primeiro APP

Por padrão, vamos criar nosso primeiro Hello World!

- Abrindo o Android Studio > New Project > Veremos inúmeros templates

- Por hora, usaremos um template vazio (empty), para entender o sistema do zero

- <i style="color:red"> Observação - Por padrão o Android Studio, recomenada o pacote ".example", você até pode usar, mas não pode publica-lo na loja ou algo similar, com este nome de pacote </i>

- É recomendado usar o nome do dominio da empresa, no pacote, exemplo: "br.com.fiap.nomeapp"

- Por padrão o arquivo é guardado em : C:\Users\nomeUsario\AndroidStudioProjects\HelloWorld

- Minimum SDK, é a versão do Android, lembrando que quanto mais antiga a versão, maior a compatibilidade com os dispositivos atuais do mercado, por tanto temos uma primeira escolha a fazer: <i style="color:red">maior compatibilidade</i> ou <i style="color:green">tecnologia mais nova</i> ?.. nesse primiro projeto usaremos a API 27 (Oreo)!

- Quando abrimos, ele inicia o carregamento do Gradle (Ferramenta que automatica todo o processo de construção do projeto)

- Após o carregamento automatico, temos a seguinte estrutura:

<center>



</center>

Sendo :

- <b>Manifest</b>, com o AndroidManifest.xml, responsável pelo gerenciamento de permissões, config do projeto, como estilo da activity, qual o nome do app, qual será a principal activity, o ícone e outras functions que veremos mais tarde.

- <b>Java</b>, com Três subspastas: a primeira, são as classes do projeto, na segunda (androidtTest), pode ser elaborado o teste automatizado da nossa aplicação, e na terceira (test), criaremos o teste de interface gráfica.

- <b>RES</b>, com 4 subpastar, drawble, na qual haverá imagéns dimensões fixas, a segunda, o mipmap, na qual haverá imagen que se adaptam de acordo com o dispositivo, a terceira, values, na qual criam-se XML padrões de cores, strings, themes, facilitando a reutilização no projeto.

## Executar projeto em device Físico via USB

Para isso é necessário habilitas, nas configurações do aparelho em:

- Configurações> Segurança> Fontes Desconhecidas e a opção Configurações > Opções de desenvolvedor> Depuração USB 

## Opção Ocultando no Android 3.2 ou superior 

- A opção de desenvolvedor não aparece por padrão no Android 3.2 ou superior, para habilitá-la é preciso entrarem Configurações> Sobre o Telefone > Informações do Software e clicar 7x em Número de compilação, ao fazer isso, uma mensagem vai aparecer informando que agora você é desenvolvedor e o menu Opções de desenvolvedor estará habilitado.
