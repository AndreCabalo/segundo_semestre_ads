## Cultura Startup

Alguma das caracterstica comuns da cultura Statup:

- Adaptabilidade

- Flexibilidade

- Erro como aprendizado

- Menos hierarquia e mais autonomia

- Otimizao de recursos

## Como funcionar o Startup One?

- Teremos Lives e encontros especficos para os alunos que optarem pela Startup One

- Durante o ano, iremos apernder na prtica:

        - Identificar oportunidades;

        - Validar hipteses do problema;

        - Desenhar propostas de valor e modelos de negcio com o Canvas;

        - Prototipar;

        - Validar hipteses da sua soluo;

        - Analisar a viabilidade financeira do seu projeto;
        
        - Aprensentar o seu projeto, atravs do Pitch;




## Premiao

Aqueles que chegarem mais prximos de um negcio e tiverem maior nvel de maturidade:

- Iro participar de uma competio, fazendo pitch para uma banca de investidores, aceleradoras e hubs de inovao de grandes empresas.

- A Startup vencedora ganhar uma extenso internacional, totalmente gratuita, na maior escola de negcio do mundo: Babson College, em Boston nos EUA.
