# Projeto Startup One

Na FIAP, substituimos o TCC por um projeto Startup, onde vc construir um negocio, no qual você põem literalmente a mão na massa e sai com catalogo/negocio que pode lhe gerar lucros.

<h2 style="color:springgreen">Passos:</h2>

1 - Identificar oportunidades

2 - Validar problemas

3 - Proposta de valor / Canvas modelo de negocio

4 - Prototipação

5 - Validar solução

6 - Financeiro

7 - Pitch, apresentar Startup

## Cultura Startup

Alguma das característica comuns da cultura Statup:

- Adaptabilidade

- Flexibilidade

- Erro como aprendizado

- Menos hierarquia e mais autonomia

- Otimização de recursos

## Como funcionará o Startup One?

- Teremos Lives e encontros específicos para os alunos que optarem pela Startup One

- Durante o ano, iremos apernder na prática:

        - Identificar oportunidades;

        - Validar hipóteses do problema;

        - Desenhar propostas de valor e modelos de negócio com o Canvas;

        - Prototipar;

        - Validar hipóteses da sua solução;

        - Analisar a viabilidade financeira do seu projeto;
        
        - Aprensentar o seu projeto, através do Pitch;




## Premiação

Aqueles que chegarem mais próximos de um negócio e tiverem maior nível de maturidade:

- Irão participar de uma competição, fazendo pitch para uma banca de investidores, aceleradoras e hubs de inovação de grandes empresas.

- A Startup vencedora ganhará uma extensão internacional, totalmente gratuita, na maior escola de negócio do mundo: Babson College, em Boston nos EUA.
