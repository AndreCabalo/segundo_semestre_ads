Principais diferenças entre mobile Ios e Android:

Ambos são mais fáceis de programas do que para PC-Web

Ambas são consideravelmente parecidas, pois compartilham elementos proximos, praticamente uma foi compiando os beneficios da outra.

## Ios

- 30% dos usuários utilizam (em 2023)

- Precisa de um equipamento Apple para publicas além da exigencia da taxa de desenvolvedor para publicação $99,00/ano, para entrar na loja

### Primeiros passos com Ios

1 -  Baixar Ebook do Swift para aprender o basico

2 - Usar um software (Por exemplo: REPL - Read-Eval-Print Loop) para entender a dinamica da linguagem,e testar para criar classse, chamar o método 

3 - Se gostar, utilize um Mac na MacCloud.com, onde tera uma máquina mac, para usar o XCode

4 - Interagir com varias telas, interagir model com controler e view

5 - Acesso a Json

6 - Persistencia de dados

## Android

- Taxa para confirmação de que é um ser humano de R$ 20,00 para confirmaçãoo do desenvolvedor

- ~70% dos usuários (em 2023)

### Primeiro passos com Android:

- Baixar o Android Studio (um set de ferramentas), apos isso:

    1 - Foque em aprender os componentes visuais

    2 - Brinque com responsividade de telas

    3 - Comunicação com API usuando HTTPD

    4 - Jetpack

    5 - Acessibilidade

    6 - ProGuard

Apos isso esta pronto para fazer 99% dos Apps
